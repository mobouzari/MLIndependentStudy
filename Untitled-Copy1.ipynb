```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
import sklearn
import pandas
```

```python
df = pandas.read_csv('contactid_08_01_18download.csv', encoding="utf-8")
```

```python
df
```

```python
dfclean = df.loc[df['cat'] == 'Phone survey']
```

```python
dfclean
```

```python
dfclean = df.loc[df['close'].notnull()]
```

```python
dfclean
```

```python
x = dfclean[['Total number of calls', 'Total duration of calls']].values #,'Total incoming calls','Total outgoing calls','Proportion of incoming calls answered','Proportion of calls incoming','Duration of incoming calls','Duration of outgoing calls','Total number of night calls']].values
```

```python
x
```

```python
x = x.notnull()
```

```python
x.shape
```

```python
# ravel creates 1d array
y = dfclean[['close']].values.ravel()
y
```

```python
def transform(y):
    if y < 33:
        return 0
    if y < 66:
        return 1
    if y < 100:
        return 2
    return 3

yp = [transform(i) for i in y]
```

```python
#followed original mock-up
from sklearn.ensemble import GradientBoostingClassifier
classifier = GradientBoostingClassifier()
```

```python
classifier.fit(x,yp)
```

```python
yhat = classifier.predict(x)
```

```python
yhat
```

```python
accuracy_score(yp, yhat)
```

```python
from sklearn import svm
regressor = svm.SVR()
regressor.fit(x,y)
```

```python
yhat = regressor.predict(x)
yhat
```

```python
mean_squared_error(y, yhat)
```
